```python
class Batch_Regressor:
    
    def __init__(self,lr=0.5,epochs=1000):
        self.coef=None
        self.intercept=None
        self.lr=lr
        self.epochs=epochs
        

    def fit(self,x_train,y_train):
        # starting value of b0=0
        self.intercept=0
        self.coef=np.zeros(x_train.shape[1])

        for i in range(self.epochs):
            # find value of y_hat
            self.y_hat=self.intercept+np.dot(x_train,self.coef)
            # find derivatives
            self.coef_der = -2 * np.dot(x_train.T, (y_train - self.y_hat)) / x_train.shape[0]
            self.coef=self.coef-(self.lr* self.coef_der)
            self.intercept_der = -2 * np.mean(y_train - self.y_hat)
            self.intercept=self.intercept-(self.lr*self.intercept_der)
            
        print(self.coef,self.intercept)
        
    def predict(self,x_test):
        return np.dot(x_test,self.coef) + self.intercept
        
```

```python
br= Batch_Regressor()
br.fit(x_train,y_train)
```

```python
y_pred1= br.predict(x_test)
print(y_pred1)
```

```python
print(r2_score(y_test,y_pred1))
```
