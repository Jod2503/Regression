```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x,y=load_diabetes(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
import random
class Mini_batch_Regressor:
    
    def __init__(self,batch_size,lr,epochs):
        self.coef=None
        self.intercept=None
        self.lr=lr
        self.epochs=epochs
        self.batch_size=batch_size
        

    def fit(self,x_train,y_train):
        # starting value of b0=0
        self.intercept=0
        self.coef=np.ones(x_train.shape[1])

        for i in range(self.epochs):
             for j in range(int(x_train.shape[0]/self.batch_size)):
                 idx=random.sample(range(0,x_train.shape[0]),self.batch_size)
                 y_hat=np.dot(x_train[idx],self.coef)+ self.intercept
                 # as only one row, no need of summation and n
                 intercept_der=-2*(np.mean(y_train[idx]-y_hat))
                 self.intercept=self.intercept-(self.lr*intercept_der)
                 coef_der=-2* (np.dot((y_train[idx]-y_hat),x_train[idx]))
                 self.coef=self.coef-(self.lr*coef_der)
                 
        print(self.coef,self.intercept)
        print(self.coef.shape)
        print(self.intercept.shape)
        # coef should be a vector but intercept should be a scaler(as only beta not)

            
    def predict(self,x_test):
        return(np.dot(x_test,self.coef)+self.intercept)
        
```

```python
import random
random.sample(range(0,100),10)
```

```python
mb=Mini_batch_Regressor(35,0.01,1000)
mb.fit(x_train,y_train)
y_pred=mb.predict(x_test)
```
