```python
import numpy as np
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x,y=load_diabetes(return_X_y=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
class Stochastic_Regressor:
    
    def __init__(self,lr,epochs):
        self.coef=None
        self.intercept=None
        self.lr=lr
        self.epochs=epochs
        

    def fit(self,x_train,y_train):
        # starting value of b0=0
        self.intercept=0
        self.coef=np.ones(x_train.shape[1])

        for i in range(self.epochs):
             for j in range(x_train.shape[0]):
                 # we select a random index from 0 to 353
                 # loop utni baar chalega jitne rows honge
                 idx=np.random.randint(0,x_train.shape[0])
                 y_hat=np.dot(x_train[idx],self.coef)+ self.intercept
                 # as only one row, no need of summation and n
                 intercept_der=-2* (y_train[idx]-y_hat)
                 self.intercept=self.intercept-(self.lr*intercept_der)
                 coef_der=-2* (np.dot((y_train[idx]-y_hat),x_train[idx]))
                 self.coef=self.coef-(self.lr*coef_der)
                 
        print(self.coef,self.intercept)
        print(self.coef.shape)
        print(self.intercept.shape)
        # coef should be a vector but intercept should be a scaler(as only beta not)

            
    def predict(self,x_test):
        return(np.dot(x_test,self.coef)+self.intercept)
        
```

```python
sgd=Stochastic_Regressor(0.01,50)
sgd.fit(x_train,y_train)
y_pred=sgd.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
